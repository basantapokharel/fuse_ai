Step 1: Import Libraries & Load Dataset

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load tips dataset
df = sns.load_dataset("tips")
df.head()
```

total_bill: Total amount of the bill.

tip: Tip given.

sex: Gender of the person who paid.

smoker: Whether the group included smokers.

day: Day of the week.

time: Time of day (Lunch or Dinner).

size: Number of people in the party.

Step 3: Univariate Analysis (One variable)

```python
#Distribution of total bill
sns.histplot(df['total_bill'], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Countplot for gender
sns.countplot(x='sex', data=df)
plt.title("Gender Count")
plt.show()
```

```python
# Countplot for smoker
sns.countplot(x='smoker', data=df)
plt.title("Smoker vs. Non-smoker")
plt.show()
```

 Step 4: Bivariate Analysis (Two variables)

```python
# Scatter plot: Total bill vs Tip
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# Boxplot: Tip by Gender
sns.boxplot(x='sex', y='tip', data=df)
plt.title("Tip by Gender")
plt.show()
```

```python
# Boxplot: Total bill by day
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Total Bill by Day")
plt.show()
```

Step 5: Multivariate Analysis (Three or more variables)

```python
# Pairplot with hue
sns.pairplot(df, hue='sex')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
# Heatmap of correlations
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Plot heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Step 6: Key Insights

1) Customers tend to tip more when their total bill is higher, although the relationship is not strictly linear.

2. Male customers tend to tip slightly more than females, especially for higher bills.

3. Smokers and non-smokers show different tipping behavior, with smokers tipping slightly less on average.

4. Saturday and Sunday are the busiest days, with higher total bills.
